```python

#1. Assess the overall distribution of the data to Male vs. Female clients, as well as how their age and gender differ.
#2. Evaluate which gender has a higher number of account opening.
#3. Elvaluate which age group has the most account openings.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

```

```python
clients = pd.read_csv('C:\data\Traindata.csv')
```

```python
clients.head()
```

```python
clients.describe()
```

```python
# Men and women are equally in the data set.
sns.countplot(x='Gender', data=clients);
plt.title('Distribution of Gender');


```

```python
sns.countplot(x='Age',data=clients);
#sns.set(rc = {'figure.figsize':(20,10)})
plt.title('Distribution by frequency of opening an accounts');
```

```python
clients.hist('Age', bins=35);
plt.title('Distribution of Age');
plt.xlabel('Age');
```

```python
# Age histogram by gender
#Based on the data set there are more women than men, aslo there are lot of men in the middle age.



plt.hist('Age', data=clients[clients['Gender'] == 'FEMALE'], alpha=0.5, label='Male');
plt.hist('Age', data=clients[clients['Gender'] == 'MALE'], alpha=0.5, label='Female');
plt.title('Distribution of Age by Gender');
plt.xlabel('Age');
plt.legend();
```

```python
# Histogram of income by gender
plt.hist('Monthly Income(1-100 K)', data=clients[clients['Gender'] == 'MALE'], alpha=0.5, label='Male');
plt.hist('Monthly Income(1-100 K)', data=clients[clients['Gender'] == 'FEMALE'], alpha=0.5, label='Female');
plt.title('Distribution of Income by Gender');
plt.xlabel('Income (Thousands of Pesos)');
plt.legend();
```

```python
sns.heatmap(clients.corr(), annot=True)
plt.title('Correlation Heatmap-to all variables');
```

```python
# Create data sets by gender
male_clients = clients[clients['Gender'] == 'MALE']
female_clients = clients[clients['Gender'] == 'FEMALE']
```

```python
# Based on the data set , women open accounts at a slightly higher in 1 points than men.

print(male_clients['Monthly Income(1-100 K)'].mean())
print(female_clients['Monthly Income(1-100 K)'].mean())
```

```python
sns.heatmap(female_clients.corr(), annot=True);
plt.title('Correlation Heatmap - Female');
```

```python
sns.heatmap(male_clients.corr(), annot=True);
plt.title('Correlation Heatmap - Male');
```
